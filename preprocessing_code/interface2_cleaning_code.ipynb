```python
import pandas as pd
import zipfile
```

```python
zf = zipfile.ZipFile('./data/U.S._Chronic_Disease_Indicators__CDI.zip')
chronic_disease_df = pd.read_csv(zf.open('U.S._Chronic_Disease_Indicators__CDI_.csv'))
```

```python
questions = ['Asthma mortality rate',
             'Cancer of the oral cavity and pharynx, mortality',
             'Cancer of the colon and rectum (colorectal), mortality',
             'Cancer of the lung and bronchus, mortality',
             'Cancer of the female cervix, mortality'
             'Melanoma, mortality',
             'Mortality with end-stage renal disease',
             'Mortality from cerebrovascular disease (stroke)',
             'Mortality from coronary heart disease',
             'Chronic liver disease mortality',
             'Obesity among adults aged >= 18 years',
             'Healthy weight among adults aged >= 18 years',
             'Per capita alcohol consumption among persons aged >= 14 years',
             'Bringe drinking prevalence among adults aged >= 18 years',
             'Current smoking among adults aged >= 18 years']
chronic_disease_df_new = chronic_disease_df.loc[chronic_disease_df['Question'].isin(questions)]

chronic_disease_df_new = chronic_disease_df_new.copy()
chronic_disease_df_new.loc[chronic_disease_df_new.Question == 'Asthma mortality rate', 'Question'] = 'Asthma'
chronic_disease_df_new.loc[chronic_disease_df_new.Question == 'Cancer of the oral cavity and pharynx, mortality', 'Question'] = 'Oral cavity and pharynx cancer'
chronic_disease_df_new.loc[chronic_disease_df_new.Question == 'Cancer of the lung and bronchus, mortality', 'Question'] = 'Lung and bronchus cancer'
chronic_disease_df_new.loc[chronic_disease_df_new.Question == 'Cancer of the female cervix, mortality', 'Question'] = 'Female cervix cancer'
chronic_disease_df_new.loc[chronic_disease_df_new.Question == 'Melanoma, mortality', 'Question'] = 'Melanoma'
chronic_disease_df_new.loc[chronic_disease_df_new.Question == 'Mortality with end-stage renal disease', 'Question'] = 'End-stage renal disease'
chronic_disease_df_new.loc[chronic_disease_df_new.Question == 'Mortality from cerebrovascular disease (stroke)', 'Question'] = 'Stroke'
chronic_disease_df_new.loc[chronic_disease_df_new.Question == 'Mortality from coronary heart disease', 'Question'] = 'Coronary heart disease'
chronic_disease_df_new.loc[chronic_disease_df_new.Question == 'Chronic liver disease mortality', 'Question'] = 'Chronic liver disease'
chronic_disease_df_new.loc[chronic_disease_df_new.Question == 'Obesity among adults aged >= 18 years', 'Question'] = 'Obesity'
chronic_disease_df_new.loc[chronic_disease_df_new.Question == 'Healthy weight among adults aged >= 18 years', 'Question'] = 'Healthy weight'
chronic_disease_df_new.loc[chronic_disease_df_new.Question == 'Per capita alcohol consumption among persons aged >= 14 years', 'Question'] = 'Per capita alcohol consumption'
chronic_disease_df_new.loc[chronic_disease_df_new.Question == 'Bringe drinking prevalence among adults aged 18 years', 'Question'] = 'Bringe drinking prevalence'
chronic_disease_df_new.loc[chronic_disease_df_new.Question == 'Current smoking among adults aged >= 18 years', 'Question'] = 'Smoking'

chronic_disease_df_new = chronic_disease_df_new[chronic_disease_df_new['DataValueType'].isin(['Age-adjusted Rate', 'Average Annual Age-adjusted Rate'])]
chronic_disease_df_new = chronic_disease_df_new.rename(columns={'DataValue': 'Rate'})
```

```python
chronic_disease_df_new.drop('StratificationCategoryID2', axis = 1, inplace = True)
chronic_disease_df_new.drop('StratificationCategoryID3', axis = 1, inplace = True)
chronic_disease_df_new.drop('StratificationID2', axis = 1, inplace = True)
chronic_disease_df_new.drop('StratificationID3', axis = 1, inplace = True)
```

```python
interface_2_df = chronic_disease_df_new[['YearStart', 'LocationDesc', 'Question', 'LocationID', 'Rate','StratificationCategory1', 'Stratification1', 'StratificationCategoryID1', 'StratificationID1']]
```
